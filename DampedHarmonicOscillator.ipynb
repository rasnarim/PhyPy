## üõ†Ô∏è Simple Harmonic Oscillator (SHO) üåÄ

Before diving into damped oscillations, let's first explore the **Simple Harmonic Oscillator (SHO)**, a system where the restoring force is directly proportional to displacement.

### ‚öôÔ∏è The Equation of Motion:
The motion of a simple harmonic oscillator is described by the differential equation:

$$ m \ddot{x} + kx = 0 $$

where:
- $m$ is the mass,
- $k$ is the spring constant,
- $x(t)$ is the displacement at time $t$.

The system undergoes sinusoidal motion, with the displacement $x(t)$ given by:

$$ x(t) = A \cos(\omega_0 t + \phi) $$

where:
- $A$ is the amplitude,
- $\omega_0 = \sqrt{\frac{k}{m}}$ is the natural frequency of the oscillator,
- $\phi$ is the phase constant, determined by initial conditions.

### üîß Numerical Solution Using Euler's Method

In the next section, we'll solve this equation numerically using **Euler's Method** to approximate the system‚Äôs motion.


## üõ†Ô∏è Numerical Solution of the Simple Harmonic Oscillator (SHO) using Euler's Method üßÆ

To solve the SHO numerically, we rewrite the second-order differential equation as two first-order equations:

$$
\dot{x} = v
$$
$$
\dot{v} = -\frac{k}{m} x
$$

### ‚úèÔ∏è Euler's Method
Euler's method is a simple numerical technique that updates the position and velocity at each time step $\Delta t$. The update rules are:

$$
x_{n+1} = x_n + v_n \Delta t
$$
$$
v_{n+1} = v_n - \frac{k}{m} x_n \Delta t
$$

This allows us to approximate the motion of the SHO step by step. In the next section, we‚Äôll implement this method in Python and visualize the results. 


## üßë‚Äçüíª Python Code for Simple Harmonic Oscillator (SHO) üöÄ

Let's implement the **Euler Method** to simulate the simple harmonic oscillator in Python. We'll use the following parameters:
- Mass ($m$)
- Spring constant ($k$)
- Initial displacement ($x_0$)
- Initial velocity ($v_0$)

We will simulate the motion over time and visualize the displacement of the oscillator.

### üîß Code Implementation

You can run the following Python code to observe the motion of the SHO:



```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
m = 1.0  # mass (kg)
k = 1.0  # spring constant (N/m)

# Initial conditions
x0 = 1.0  # initial displacement (m)
v0 = 0.0  # initial velocity (m/s)

# Time parameters
t_max = 20  # total time (s)
dt = 0.01  # time step (s)
n_steps = int(t_max / dt)  # number of steps

# Initialize arrays for time, displacement, and velocity
t = np.linspace(0, t_max, n_steps)
x = np.zeros(n_steps)
v = np.zeros(n_steps)

# Set initial conditions
x[0] = x0
v[0] = v0

# Euler method loop to calculate the motion
for i in range(1, n_steps):
    v[i] = v[i-1] - (k/m) * x[i-1] * dt  # velocity update
    x[i] = x[i-1] + v[i-1] * dt  # position update

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(t, x, label="SHO: x(t)")
plt.title("Simple Harmonic Oscillator (SHO) - Euler Method")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.grid(True)
plt.legend()
plt.show()
```

## üîÑ Damped Harmonic Oscillator üîß

Now that we have explored the **Simple Harmonic Oscillator**, let‚Äôs move on to the **Damped Harmonic Oscillator**.

In real-world systems, energy is often lost due to resistive forces, such as friction or air resistance. This introduces damping, which causes the amplitude of oscillation to decrease over time.

### ‚öôÔ∏è The Equation of Motion:
The damped harmonic oscillator is described by the following differential equation:

$$ m \ddot{x} + b \dot{x} + kx = 0 $$

where:
- $m$ is the mass,
- $b$ is the damping coefficient,
- $k$ is the spring constant,
- $x(t)$ is the displacement.

### üîç Types of Damping:
1. **Underdamped ($b < 2\sqrt{km}$)**: The system oscillates, but the amplitude gradually decreases over time.
2. **Critically Damped ($b = 2\sqrt{km}$)**: The system returns to equilibrium as quickly as possible without oscillating.
3. **Overdamped ($b > 2\sqrt{km}$)**: The system slowly returns to equilibrium without oscillating.

In the next sections, we'll explore the numerical solution and visualize different damping behaviors.



## üõ†Ô∏è Numerical Solution of the Damped Harmonic Oscillator üíª

To solve the **damped harmonic oscillator** numerically, we will use the same approach as with the simple harmonic oscillator by breaking it down into first-order differential equations.

The system of equations becomes:

$$ \dot{x} = v $$
$$ \dot{v} = -\frac{k}{m} x - \frac{b}{m} v $$

### ‚úèÔ∏è Euler‚Äôs Method for Damped Oscillator
We use **Euler's method** again to update the position and velocity step-by-step. The update rules are:

$$ x_{n+1} = x_n + v_n \Delta t $$
$$ v_{n+1} = v_n + \left(-\frac{k}{m} x_n - \frac{b}{m} v_n\right) \Delta t $$

In the next section, we will implement this numerically and plot the motion for different values of the damping coefficient $b$.


## üõ†Ô∏è Implementation: Damped Harmonic Oscillator üìä

Now let‚Äôs numerically solve the damped harmonic oscillator for one specific damping case using the Euler method. We'll start with a **damping coefficient $b = 0.5$**, which represents an **underdamped system**.

### üìù Python Code for One Oscillator:



```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
m = 1.0  # mass (kg)
k = 1.0  # spring constant (N/m)
b = 0.5  # damping coefficient (N‚ãÖs/m)

# Initial conditions
x0 = 1.0  # initial displacement (m)
v0 = 0.0  # initial velocity (m/s)

# Time parameters
t_max = 20
dt = 0.01
n_steps = int(t_max / dt)

# Initialize arrays for time, displacement, and velocity
t = np.linspace(0, t_max, n_steps)
x = np.zeros(n_steps)
v = np.zeros(n_steps)

# Set initial values
x[0] = x0
v[0] = v0

# Euler method loop to calculate damped oscillator motion
for i in range(1, n_steps):
    v[i] = v[i-1] + (-k/m * x[i-1] - b/m * v[i-1]) * dt
    x[i] = x[i-1] + v[i-1] * dt

# Plot the result
plt.figure(figsize=(10, 6))
plt.plot(t, x, label=f'Damped Oscillator (b = {b})')
plt.title("Damped Harmonic Oscillator - Underdamped Case")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.grid(True)
plt.legend()
plt.show()
```

## üõ†Ô∏è Critically Damped Oscillator ‚öñÔ∏è

Now let‚Äôs explore the **critically damped** situation, which occurs when the damping coefficient $b$ is just enough to prevent oscillations but allows the system to return to equilibrium as quickly as possible.

### ‚öôÔ∏è The Critical Damping Condition
The critical damping coefficient is given by:

$$ b_{\text{critical}} = 2\sqrt{km} $$

For a system where $m = 1.0$ kg and $k = 1.0$ N/m, the critical damping coefficient is:

$$ b_{\text{critical}} = 2\sqrt{1.0 \cdot 1.0} = 2.0 $$

In this case, the system will return to equilibrium without oscillating. 

---

### üìù Python Code for Multiple Damping Cases

We will now simulate the motion of the damped oscillator for different values of $b$, including:
- **Underdamped** ($b = 0.5$)
- **Critically Damped** ($b = 2.0$)
- **Overdamped** ($b = 2.5$)


```python
# Define constants
m = 1.0  # mass (kg)
k = 1.0  # spring constant (N/m)
b_values = [0.5, 2.0, 2.5]  # damping coefficients for underdamped, critical, and overdamped

# Initial conditions
x0 = 1.0  # initial displacement (m)
v0 = 0.0  # initial velocity (m/s)

# Time parameters
t_max = 20
dt = 0.01
n_steps = int(t_max / dt)

# Initialize time array
t = np.linspace(0, t_max, n_steps)

# Loop over different damping values
plt.figure(figsize=(10, 6))
for b in b_values:
    # Initialize arrays for displacement and velocity
    x = np.zeros(n_steps)
    v = np.zeros(n_steps)

    # Set initial conditions
    x[0] = x0
    v[0] = v0

    # Euler method loop for damped oscillator motion
    for i in range(1, n_steps):
        v[i] = v[i-1] + (-k/m * x[i-1] - b/m * v[i-1]) * dt
        x[i] = x[i-1] + v[i-1] * dt

    # Plot the result for each damping case
    plt.plot(t, x, label=f'b = {b}')

# Plot configuration
plt.title("Damped Harmonic Oscillator for Different Damping Coefficients")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.grid(True)
plt.legend()
plt.show()
```

In this exercise, we simulate a **damped harmonic oscillator** where the damping coefficient $b$ suddenly becomes zero at a time $t_{\text{failure}}$. This could represent a system where the damping force (such as friction or resistance) is suddenly removed.

### Tasks:

2. **Remove Damping at $t_{\text{failure}}$**: At time $t_{\text{failure}}$, set $b = 0$ and continue the simulation.
3. **Plot and Analyze**: Observe how the system behaves before and after the damping fails.

**Hint**: You'll need to modify the Euler method to change $b$ after $t_{\text{failure}}$.

### üîß Initial Values for the Exercise:

To get started, use the following initial conditions and parameters for the system:

- **Mass**: $m = 1.0$ kg
- **Spring constant**: $k = 1.0$ N/m
- **Initial displacement**: $x_0 = 1.0$ m
- **Initial velocity**: $v_0 = 0.0$ m/s
- **Damping coefficient** (before failure): $b = 0.5$
- **Time of damping failure**: $t_{\text{failure}} = 10$ s
- **Simulation time**: $t_{\text{max}} = 20$ s

These values will allow you to observe the system's behavior before and after the damping is removed.



## üí° Answer to the Exercise: Oscillator with Sudden Loss of Damping

### Explanation:

1. **Before Damping Failure ($t < t_{\text{failure}}$)**:
   - The system exhibits **underdamped** motion with the damping coefficient $b = 0.5$. The oscillations occur with a gradually decreasing amplitude due to energy loss from damping.

2. **After Damping Failure ($t \geq t_{\text{failure}}$)**:
   - When $b = 0$ at $t_{\text{failure}} = 10$ s, the system transitions to a **simple harmonic oscillator** with no damping. The oscillations continue indefinitely with constant amplitude, as there's no energy loss.

### Python Code:

```python
# Define constants
m = 1.0  # mass (kg)
k = 1.0  # spring constant (N/m)
b_initial = 0.5  # damping coefficient before failure (N‚ãÖs/m)
t_failure = 10.0  # time when damping fails (s)

# Initial conditions
x0 = 1.0  # initial displacement (m)
v0 = 0.0  # initial velocity (m/s)

# Time parameters
t_max = 20
dt = 0.01
n_steps = int(t_max / dt)

# Initialize arrays for time, displacement, and velocity
t = np.linspace(0, t_max, n_steps)
x = np.zeros(n_steps)
v = np.zeros(n_steps)

# Set initial values
x[0] = x0
v[0] = v0

# Euler method loop with damping failure at t_failure
for i in range(1, n_steps):
    # Apply damping before t_failure
    if t[i] < t_failure:
        b = b_initial
    else:
        b = 0.0  # no damping after failure

    # Euler method updates
    v[i] = v[i-1] + (-k/m * x[i-1] - b/m * v[i-1]) * dt
    x[i] = x[i-1] + v[i-1] * dt

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(t, x, label="Damped Oscillator with Damping Failure")
plt.axvline(x=t_failure, color='r', linestyle='--', label=f't_failure = {t_failure}s')
plt.title("Oscillator with Damping Failure")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.legend()
plt.grid(True)
plt.show()
```
